```python
import pypoet as poet
from matplotlib import pyplot as plt
```

```python
#hamilton_score = poet.full_score(poet.text.hamilton)
stackoverflow_score = poet.full_score(poet.text.stackoverflow)
nyt_score = poet.full_score(poet.text.nyt)
aerobics_score = poet.full_score(poet.text.aerobics)
```

```python
#plt.plot(hamilton_score, color='r')
plt.plot(stackoverflow_score, color='g')
plt.plot(nyt_score, color='b')
plt.plot(aerobics_score, color='k')
plt.show()
```

```python
scores = [
    #hamilton_score,
    stackoverflow_score,
    nyt_score,
    aerobics_score]

min_length = min([len(score) for score in scores])
average_score = sum([score[:min_length] for score in scores]) / len(scores)
#plt.plot(hamilton_score - average_score, color='r')
plt.plot(stackoverflow_score[:min_length] - average_score, color='g')
plt.plot(nyt_score[:min_length] - average_score, color='b')
plt.plot(aerobics_score[:min_length] - average_score, color='k')
plt.show()
```
