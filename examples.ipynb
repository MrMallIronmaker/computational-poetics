```python
import rap_metrics as rm
from matplotlib import pyplot as plt
```

```python
hamilton_score_0 = rm.full_score(rm.hamiltons[0])
hamilton_score_1 = rm.full_score(rm.hamiltons[1])
stackoverflow_score = rm.full_score(rm.stackoverflow)
nyt_score = rm.full_score(rm.nyt)
```

```python
plt.plot(hamilton_score_0, color='r')
plt.plot(hamilton_score_1, color='g')
plt.plot(stackoverflow_score, color='b')
plt.plot(nyt_score, color='k')
plt.show()
```

```python
average_score = (hamilton_score_0 + hamilton_score_1 +
                 stackoverflow_score + nyt_score) / 4.0
plt.plot(hamilton_score_0 - average_score, color='r')
plt.plot(hamilton_score_1 - average_score, color='g')
plt.plot(stackoverflow_score - average_score, color='b')
plt.plot(nyt_score - average_score, color='k')
plt.show()
```
