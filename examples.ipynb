```python
import pypoet as poet
from matplotlib import pyplot as plt
```

```python
split_hamilton = poet.text.hamilton.split('\n')
hamilton_0 = '\n'.join(split_hamilton[0:25])
hamilton_1 = '\n'.join(split_hamilton[25:50])
```

```python
texts, names = zip(*[
    (hamilton_0, "ham 0"),
    (hamilton_1, "ham 1"),
    (poet.text.stackoverflow, "stack"),
    (poet.text.nyt, "nyt"),
    (poet.text.aerobics, "aero"),
    (poet.text.lrh, "lrh")
])
scores = [poet.full_score(text) for text in texts]
colors = "rgbkycm"
```

```python
fig = plt.figure()
ax = plt.subplot(111)

for (name, score, color) in zip(names, scores, colors):
    ax.plot(score, color=color, label=name)
ax.legend()
plt.show()
```

```python
min_length = min([len(score) for score in scores])
average_score = sum([score[:min_length] for score in scores]) / len(scores)

fig = plt.figure()
ax = plt.subplot(111)

for (name, score, color) in zip(names, scores, colors):
    ax.plot(score[:min_length] - average_score, color=color, label=name)
ax.legend()
plt.show()
```
