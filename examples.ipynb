```python
import pypoet as poet
from matplotlib import pyplot as plt
```

```python
hamilton_score = poet.full_score(poet.text.hamilton)
stackoverflow_score = poet.full_score(poet.text.stackoverflow)
nyt_score = poet.full_score(poet.text.nyt)
```

```python
plt.plot(hamilton_score, color='r')
plt.plot(stackoverflow_score, color='g')
plt.plot(nyt_score, color='b')
plt.show()
```

```python
average_score = (hamilton_score + stackoverflow_score + nyt_score) / 3.0
plt.plot(hamilton_score - average_score, color='r')
plt.plot(stackoverflow_score - average_score, color='g')
plt.plot(nyt_score - average_score, color='b')
plt.show()
```
